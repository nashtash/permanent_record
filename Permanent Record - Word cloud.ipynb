```python
import nltk
import string
from nltk.corpus import stopwords
 
with open('Edward Snowden - Permanent Record-Metropolitan Books (2019).txt','r') as inFile, open('outputfile.txt','w') as outFile:
    for line in inFile.readlines():
        print(" ".join([word for word in line.lower().translate(str.maketrans('', '', string.punctuation)).split()
                        if len(word) >=4 and word not in stopwords.words('english')]), file=outFile)
```

```python
from wordcloud import WordCloud
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (16.0, 9.0)
```

```python
script = open("outputfile.txt").read()


wc = WordCloud(background_color="white",
               width=1600, height=900, colormap=matplotlib.cm.inferno)

wc.generate(script)

plt.figure()
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
```

```python
from PIL import Image
import numpy as np
from wordcloud import ImageColorGenerator
```

```python
mask = np.array(Image.open("9781250237231_custom-ac0f29c2f815d79e03c3157d988ae95f9649f3be-s300-c85.jpg"))

wc = WordCloud(background_color="white",
               width=1280, height=628, mask=mask)
wc.generate(script)

image_colors = ImageColorGenerator(mask)

plt.figure()
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
```
